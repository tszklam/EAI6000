Please differentate __propositional logic__, __first order logic__, and __probability theory__ their roles in knowledge representation

__Propositional Logic__:

Propositional logic is a branch of logic that deals with propositions (statements that are either true or false, 1 or 0) and the relationships between them. It is the simplest form of logic and is primarily concerned with the logical connectives (such as AND, OR, NOT, IMPLIES, etc.) that can be used to build complex expressions from simpler ones. In knowledge representation, propositional logic is used to represent the relationships between facts, but it may lack the expressive power to represent more complex relationships or quantify over variables.

__First-Order Logic__:

First-order logic (FOL) is an extension of propositional logic that allows for the representation of more complex relationships by introducing quantifiers (such as FOR ALL and EXISTS) and variables. In FOL, knowledge is represented using predicates, which are functions that map objects to truth values. These predicates can be combined with logical connectives and quantifiers to express more sophisticated relationships between objects. It is more expressive than propositional logic and is widely used in AI or ML, particularly in knowledge representation and automated reasoning.

__Probability Theory__:

Probability theory is a mathematical framework for representing and reasoning about uncertain things. It provides a way to quantify the odds of events, given certain conditions or evidence. In knowledge representation, probability theory can be used to model uncertain or incomplete information by associating probabilities with facts or relationships between facts. This allows for the representation of level of belief/confidence in certain propositions, rather than just binary true or false values. Probabilistic reasoning techniques, like the Bayesian networks and Markov models, can then be employed to reason with this uncertain knowledge and make informed decisions or predictions.

Please describe __Bayes Rule__ and __Naive Bayes classifier__ in your own words.

__Bayes Rule__ in statistics is a fundamental principle that provides a method for updating the probability of an event or hypothesis based on new evidence. In essence, Bayes Rule calculates the conditional probability of an event occurring given some prior knowledge or evidence. The rule states that the updated probability of an event (posterior probability) is proportional to the product of the prior probability of the event and the likelihood of the evidence given the event.

A __Naive Bayes classifier__ is a machine learning algorithm based on Bayes Rule that is used for classification tasks. It is called "naive" because it makes a strong assumption that the features used to describe an instance in the dataset are conditionally __independent__ given the class label. Despite this simplification, Naive Bayes classifiers often perform well in practice, especially in situations with limited data or when the independence assumption is reasonably accurate. To classify an instance, the Naive Bayes classifier calculates the posterior probability for each possible class label given the instance's feature values, and then selects the class with the highest probability as the predicted label.

### <div class="alert alert-info">[GRADED  TASK 2.2]</div>
Write function `wine_classification` that does the following:
* loads the wine dataset using sklearn (`sklearn.datasets.load_wine`)
* splits the data into training and testing part using the `train_test_split` function so that the training set size is 80% of the whole data (set argument `random_state=2023` to make the result deterministic, and make sure the data is split in a stratified fashion)
* use Gaussian Naive Bayes to fit the training data
* predict labels of the test data
* the function should return accuracy score as the prediction performance (`sklearn.metrics.accuracy_score`)

Please add docstring, comment on your code, and interpret the result

```python
# Load the modules in the sklearn library
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix

def wine_classification():
    """
    This function performs Gaussian Naive Bayes classification on the wine dataset, returns the accuracy score,
    and prints/plots the confusion matrix for the test data.
    """

    # Load the wine dataset
    wine_data = load_wine()
    
    X = wine_data.data
    y = wine_data.target
    
    # Split the data into training and testing sets (80% for training, 20% for testing)
    # Set random seed to 2023 to reproduce results
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2023, stratify=y)
    
    # Instantiate and fit the Gaussian Naive Bayes model to the training data
    gnb = GaussianNB()
    gnb.fit(X_train, y_train)
    
    # Make predictions on the test data
    y_pred = gnb.predict(X_test)
    
    # Lets do some plot
    import seaborn as sns
    import matplotlib.pyplot as plt

    # Do the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    print(cm)

    # Plot the confusion matrix as a heatmap
    # Define the size of canvas
    plt.figure(figsize=(8, 6))
    # Define the heatmap
    sns.heatmap(cm, annot=True, cmap="Blues", fmt='d', xticklabels=wine_data.target_names, yticklabels=wine_data.target_names)
    # Minor definations
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title('Confusion Matrix')
    plt.show()

    # Calculate and return the accuracy score
    accuracy = accuracy_score(y_test, y_pred)

    return accuracy

# Call the function and print the accuracy score
accuracy = wine_classification()
print(f"The accuracy of the Gaussian Naive Bayes classifier on the wine dataset is: {accuracy:.3f}")
```
